```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../save/AGENTGA/AGENTGA-cbench-2023-08-27-16-17-55/cbench-automotive-qsort1-result.csv')
result = df['result']
result_dict = eval(result.values[0])
r = []
for i in result_dict['result_list']:
    r.append(i['best_result'])
print(min(r))
plt.title('compile size of cbench-automotive-susan')
plt.ylabel('size')
plt.xlabel('generation')
plt.plot(r)
plt.savefig('agentga.png')
# result.values
```
